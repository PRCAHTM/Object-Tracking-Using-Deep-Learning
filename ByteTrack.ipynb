```python
yolo_model_path = "/content/models/best.pt"
detector = YOLO(yolo_model_path)

args = SimpleNamespace(
    track_thresh=0.5,
    track_buffer=30,
    mot20=False,
    match_thresh=0.8
)

tracker = BYTETracker(args, 25)

video_path = "/content/v_i2_L4qquVg0_c006_video.mp4"
output_video_path = "/content/mot_output.mp4"

cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    raise IOError("Could not open video file: " + video_path)

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = 25

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

frame_id = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_id += 1

    results = detector(frame)[0]

    bbox_xyxy = []
    for det in results.boxes:
        x1, y1, x2, y2 = det.xyxy[0].tolist()
        conf = float(det.conf[0])
        bbox_xyxy.append([x1, y1, x2, y2, conf, 1.0])

    if bbox_xyxy:
        det_tensor = torch.tensor(bbox_xyxy, dtype=torch.float32)
    else:
        det_tensor = torch.empty((0, 6), dtype=torch.float32)

    timer = Timer()
    timer.tic()
    online_targets = tracker.update(det_tensor, (frame_height, frame_width), (frame_height, frame_width))
    timer.toc()

    for track in online_targets:
        track_id = track.track_id
        x1, y1, x2, y2 = track.tlbr
        x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, f"ID {track_id}", (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    out.write(frame)


cap.release()
out.release()
print(f"✅ Multi-Object Tracking completed! Output saved at {output_video_path}")
```

```python
def clean_and_add_column_to_gt_file(input_path, output_path):
    """
    Reads the ground truth file (with 9 fields per line) and adds a placeholder (-1)
    for the missing 10th field.
    """
    import pandas as pd
    df = pd.read_csv(input_path, delimiter=',', header=None, engine='python', skipinitialspace=True)
    print("Original ground truth shape:", df.shape)

    if df.shape[1] < 10:
        num_missing = 10 - df.shape[1]
        for i in range(num_missing):
            df[df.shape[1]] = -1

    print("Cleaned ground truth shape (should be 10 columns):", df.shape)

    df.to_csv(output_path, index=False, header=False)
    print(f"Cleaned ground truth saved to: {output_path}")
    return output_path

cleaned_gt_path = "/content/cleaned_gt.txt"
clean_and_add_column_to_gt_file(gt_input, cleaned_gt_path)
```

```python
import motmetrics as mm
from tabulate import tabulate

def clean_tracking_file(file_path, output_path):
    """
    Clean the tracking results file to match the MOT format (10 fields per line).
    This function trims any extra fields and ensures correct formatting.
    """
    with open(file_path, 'r') as f:
        lines = f.readlines()

    cleaned_lines = []
    for line in lines:
        fields = line.strip().split(',')
        if len(fields) >= 10:
            cleaned_line = ','.join(fields[:10]) + '\n'
            cleaned_lines.append(cleaned_line)
        elif len(fields) < 10:
            cleaned_line = ','.join(fields + ['-1'] * (10 - len(fields))) + '\n'
            cleaned_lines.append(cleaned_line)

    with open(output_path, 'w') as f:
        f.writelines(cleaned_lines)
    print(f"Cleaned tracking results saved to: {output_path}")
    return output_path

def compute_mot_metrics(gt_file, ts_file, seqmap_file=None):
    """
    Compute MOT metrics using the motmetrics package.
    Both the ground truth and tracking results should be in MOT format.
    """
    gt = mm.io.load_motchallenge(gt_file)
    ts = mm.io.load_motchallenge(ts_file)

    acc = mm.utils.compare_to_groundtruth(gt, ts, 'iou', distth=0.5)
    mh = mm.metrics.create()
    summary = mh.compute(acc, metrics=[
        'mota', 'motp', 'idf1', 'idp', 'idr', 'precision',
        'num_false_positives', 'num_switches', 'num_fragmentations',
        'mostly_tracked', 'partially_tracked', 'mostly_lost'
    ], name='acc')
    summary_df = pd.DataFrame(summary).transpose()
    return summary_df

gt_input = "/root/.cache/kagglehub/datasets/ayushspai/sportsmot/versions/1/sportsmot_publish/dataset/val/v_G-vNjfx1GGc_c600/gt/gt.txt"


tracking_results_file = "/content/drive/My Drive/SportsMOT/test/results/tracking_results_deepsort.txt"
cleaned_results_file = "/content/drive/My Drive/SportsMOT/test/results/cleaned_tracking_results_deepsort.txt"

def clean_and_add_column_to_gt_file(input_path, output_path):
    """
    Reads the ground truth file (with 9 fields per line) and adds a placeholder (-1)
    for the missing 10th field.
    """
    df = pd.read_csv(input_path, delimiter=',', header=None, engine='python', skipinitialspace=True)
    print("Original ground truth shape:", df.shape)
    if df.shape[1] < 10:
        num_missing = 10 - df.shape[1]
        for i in range(num_missing):
            df[df.shape[1]] = -1
    print("Cleaned ground truth shape (should be 10 columns):", df.shape)
    df.to_csv(output_path, index=False, header=False)
    print(f"Cleaned ground truth saved to: {output_path}")
    return output_path

cleaned_gt_path = "/content/cleaned_gt.txt"
clean_and_add_column_to_gt_file(gt_input, cleaned_gt_path)
```

```python
def compute_mot_metrics(gt_file, ts_file, seqmap_file=None):
    """
    Computes MOT metrics using the motmetrics package.
    Both ground truth and tracking results should be in MOT format.
    """
    gt = mm.io.load_motchallenge(gt_file)
    ts = mm.io.load_motchallenge(ts_file)

    acc = mm.utils.compare_to_groundtruth(gt, ts, 'iou', distth=0.5)
    mh = mm.metrics.create()
    summary = mh.compute(acc, metrics=[
        'mota', 'motp', 'idf1', 'idp', 'idr', 'precision',
        'num_false_positives', 'num_switches', 'num_fragmentations',
        'mostly_tracked', 'partially_tracked', 'mostly_lost'
    ], name='acc')
    summary_df = pd.DataFrame(summary).transpose()
    return summary_df

gt_file_path = "/content/cleaned_gt.txt"
tracking_results_file = "/content/drive/My Drive/SportsMOT/test/results/cleaned_tracking_results_deepsort.txt"

if not os.path.exists(gt_file_path):
    print(f"Ground truth file not found: {gt_file_path}")
elif not os.path.exists(tracking_results_file):
    print(f"Tracking results file not found: {tracking_results_file}")
else:
    metrics_summary = compute_mot_metrics(gt_file_path, tracking_results_file)
    print("Raw MOT Metrics DataFrame:")
    print(metrics_summary)

    desired_metrics = [
        "mota", "precision", "idf1", "num_false_positives",
        "num_switches", "num_fragmentations", "mostly_tracked",
        "partially_tracked", "mostly_lost"
    ]

    metrics_to_print = {}
    for metric in desired_metrics:
        try:
            value = metrics_summary.loc[metric].iloc[0]
            metrics_to_print[metric] = value
        except KeyError:
            print(f"Metric '{metric}' not found in the summary.")

    metrics_to_print = {k: round(v, 4) if isinstance(v, float) else v for k, v in metrics_to_print.items()}

    table_headers = ["Metric", "Value"]
    table_rows = [[metric, metrics_to_print[metric]] for metric in desired_metrics if metric in metrics_to_print]

    print("\n+---------------------+--------------------+")
    print(tabulate(table_rows, headers=table_headers, tablefmt="grid"))
```
