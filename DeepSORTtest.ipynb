```python
import os
import cv2
import torch
import numpy as np
from pathlib import Path
from deep_sort_realtime.deepsort_tracker import DeepSort
from ultralytics import YOLO
from collections import defaultdict
from tqdm import tqdm
from typing import List, Dict, Tuple


# Constants and file paths
MODEL_PATH = Path("/content/models/model.pt")
# VIDEO_PATH = Path("/content/v_G-vNjfx1GGc_c008_video.mp4")
VIDEO_PATH = Path("/content/drive/MyDrive/DL_Project/5.mp4")
OUTPUT_VIDEO_PATH = Path("/content/drive/MyDrive/DL_Project/Out.mp4")
RESULTS_DIR = Path("/content/drive/My Drive/SportsMOT/test/results")
GROUND_TRUTH_PATH = Path("/root/.cache/kagglehub/datasets/ayushspai/sportsmot/versions/1/sportsmot_publish/dataset/val/v_ITo3sCnpw_k_c011/gt/gt.txt")

# Tracking and video settings
MIN_CONFIDENCE = 0.5
FRAME_SIZE = (1280, 720)
FPS = 25
CLASS_COLORS = {
    "player": (255, 255, 0),  # Yellow
    "referee": (0, 0, 255),   # Red
    "ball": (255, 0, 255)     # Magenta
}

# Step 1: Initialization functions

def initialize_tracker() -> DeepSort:
    return DeepSort(max_age=5, n_init=1, nms_max_overlap=0.2)

def load_yolo_model(model_path: Path) -> YOLO:
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    return YOLO(str(model_path)).to(device)

def initialize_video_capture(video_path: Path) -> cv2.VideoCapture:
    cap = cv2.VideoCapture(str(video_path))
    if not cap.isOpened():
        raise RuntimeError(f"Failed to open video {video_path}. Please check the file path and try again.")
    return cap

def initialize_video_writer(output_video_path: Path, fps: int, frame_size: Tuple[int, int]) -> cv2.VideoWriter:
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    writer = cv2.VideoWriter(str(output_video_path), fourcc, fps, frame_size)
    if not writer.isOpened():
        raise RuntimeError(f"Failed to initialize video writer for {output_video_path}. Please check the destination path.")
    return writer
```

```python

# Step 2: Data loading functions

def load_ground_truth(ground_truth_path: Path) -> Dict[int, List[Tuple[int, int, int, int]]]:
    ground_truth = defaultdict(list)
    with open(ground_truth_path, 'r') as f:
        for line in f:
            parts = line.strip().split(',')
            frame_id = int(parts[0])
            track_id = int(parts[1])
            xmin, ymin, width, height = map(int, parts[2:6])
            ground_truth[frame_id].append((track_id, xmin, ymin, width, height))
    return ground_truth
```

```python
# Step 3: Video processing functions

def process_frame_with_tracking(frame: np.ndarray, yolo_model: YOLO, tracker: DeepSort, frame_idx: int, results_file: Path) -> List[int]:
    results = yolo_model(frame, device='cuda' if torch.cuda.is_available() else 'cpu')[0]
    processed_boxes = []
    color_palette = [(0, 255, 255), (255, 0, 0), (0, 255, 0), (255, 165, 0), (255, 255, 0)]  # Color choices

    if results.boxes:
        boxes = results.boxes.xyxy.cpu().numpy()
        confidences = results.boxes.conf.cpu().numpy()

        for i, box in enumerate(boxes):
            if confidences[i] < MIN_CONFIDENCE:
                continue
            xmin, ymin, xmax, ymax = map(int, box)
            processed_boxes.append([[xmin, ymin, xmax - xmin, ymax - ymin], confidences[i]])

    tracks = tracker.update_tracks(processed_boxes, frame=frame)
    track_ids = []

    with open(results_file, "a") as f:
        for track in tracks:
            if not track.is_confirmed():
                continue
            track_id = int(track.track_id)
            track_ids.append(track_id)

            x1, y1, x2, y2 = map(int, track.to_ltrb())
            label = f"Player {track_id}"

            # Write tracking results to file
            line = f"{frame_idx + 1},{track_id},{x1},{y1},{x2 - x1},{y2 - y1},-1,-1,-1,-1\n"
            f.write(line)

            # Draw bounding box
            box_color = color_palette[track_id % len(color_palette)]
            cv2.rectangle(frame, (x1, y1), (x2, y2), box_color, 2)
            cv2.putText(frame, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, box_color, 2)

    return track_ids

def update_metrics(gt_tracks: List[Tuple[int, int, int, int]], track_ids: List[int], true_positives: int, false_positives: int, false_negatives: int) -> Tuple[int, int, int]:
    for track_id in track_ids:
        matched = False
        for gt_track in gt_tracks:
            if gt_track[0] == track_id:
                true_positives += 1
                matched = True
                break
        if not matched:
            false_positives += 1

    for gt_track in gt_tracks:
        if gt_track[0] not in track_ids:
            false_negatives += 1

    return true_positives, false_positives, false_negatives
```

```python
# Step 4: Metrics calculation function

def calculate_metrics(true_positives: int, false_positives: int, false_negatives: int, total_objects: int):
    """Calculate MOT metrics manually."""
    mota = 1 - (false_positives + false_negatives) / total_objects
    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
    idf1 = 2 * true_positives / (2 * true_positives + false_positives + false_negatives) if (2 * true_positives + false_positives + false_negatives) > 0 else 0

    print(f"MOTA: {mota:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, IDF1: {idf1:.4f}")
```

```python
import os
import cv2
import torch
import numpy as np
from pathlib import Path
from deep_sort_realtime.deepsort_tracker import DeepSort
from ultralytics import YOLO
from tqdm import tqdm
from typing import List, Tuple

# Constants and file paths
MODEL_PATH = Path("/content/drive/MyDrive/DL_Project/model.pt")
VIDEO_PATH = Path("/content/drive/MyDrive/DL_Project/5.mp4")
OUTPUT_VIDEO_PATH = Path("/content/drive/MyDrive/DL_Project/Out.mp4")
# You can change RESULTS_DIR to a directory that fits your needs.
RESULTS_DIR = Path("/content/drive/MyDrive/DL_Project/results")

# Tracking and video settings
MIN_CONFIDENCE = 0.5
FRAME_SIZE = (1280, 720)
FPS = 25
CLASS_COLORS = {
    "player": (255, 255, 0),  # Yellow
    "referee": (0, 0, 255),   # Red
    "ball": (255, 0, 255)     # Magenta
}

# Step 1: Initialization functions

def initialize_tracker() -> DeepSort:
    return DeepSort(max_age=5, n_init=1, nms_max_overlap=0.2)

def load_yolo_model(model_path: Path) -> YOLO:
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    return YOLO(str(model_path)).to(device)

def initialize_video_capture(video_path: Path) -> cv2.VideoCapture:
    cap = cv2.VideoCapture(str(video_path))
    if not cap.isOpened():
        raise RuntimeError(f"Failed to open video {video_path}. Please check the file path and try again.")
    return cap

def initialize_video_writer(output_video_path: Path, fps: int, frame_size: Tuple[int, int]) -> cv2.VideoWriter:
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    writer = cv2.VideoWriter(str(output_video_path), fourcc, fps, frame_size)
    if not writer.isOpened():
        raise RuntimeError(f"Failed to initialize video writer for {output_video_path}. Please check the destination path.")
    return writer

# Step 3: Video processing functions

def process_frame_with_tracking(frame: np.ndarray, yolo_model: YOLO, tracker: DeepSort, frame_idx: int, results_file: Path) -> List[int]:
    results = yolo_model(frame, device='cuda' if torch.cuda.is_available() else 'cpu')[0]
    processed_boxes = []
    color_palette = [(0, 255, 255), (255, 0, 0), (0, 255, 0), (255, 165, 0), (255, 255, 0)]  # Color choices

    if results.boxes:
        boxes = results.boxes.xyxy.cpu().numpy()
        confidences = results.boxes.conf.cpu().numpy()

        for i, box in enumerate(boxes):
            if confidences[i] < MIN_CONFIDENCE:
                continue
            xmin, ymin, xmax, ymax = map(int, box)
            processed_boxes.append([[xmin, ymin, xmax - xmin, ymax - ymin], confidences[i]])

    tracks = tracker.update_tracks(processed_boxes, frame=frame)
    track_ids = []

    # Ensure the parent directory exists
    results_file.parent.mkdir(parents=True, exist_ok=True)

    with open(results_file, "a") as f:
        for track in tracks:
            if not track.is_confirmed():
                continue
            track_id = int(track.track_id)
            track_ids.append(track_id)

            x1, y1, x2, y2 = map(int, track.to_ltrb())
            label = f"Player {track_id}"

            # Write tracking results to file
            line = f"{frame_idx + 1},{track_id},{x1},{y1},{x2 - x1},{y2 - y1},-1,-1,-1,-1\n"
            f.write(line)

            # Draw bounding box
            box_color = color_palette[track_id % len(color_palette)]
            cv2.rectangle(frame, (x1, y1), (x2, y2), box_color, 2)
            cv2.putText(frame, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, box_color, 2)

    return track_ids

# Step 5: Main process

def process_video():
    """Main video processing and tracking loop."""
    print("Initializing video processing...")

    # Initialize components
    tracker = initialize_tracker()
    yolo_model = load_yolo_model(MODEL_PATH)
    video_capture = initialize_video_capture(VIDEO_PATH)
    video_writer = initialize_video_writer(OUTPUT_VIDEO_PATH, FPS, FRAME_SIZE)

    total_frames = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f"Total frames in the video: {total_frames}. Starting frame processing...")

    # Process frames with tqdm progress bar
    for frame_idx in tqdm(range(total_frames), desc="Processing Frames"):
        ret, frame = video_capture.read()
        if not ret:
            print(f"Warning: Could not read frame {frame_idx + 1}. Skipping...")
            continue

        frame = cv2.resize(frame, FRAME_SIZE)
        _ = process_frame_with_tracking(frame, yolo_model, tracker, frame_idx, RESULTS_DIR / "tracking_results_deepsort.txt")
        video_writer.write(frame)

    # Release resources
    video_capture.release()
    video_writer.release()

    print("Video processing completed. Tracking results have been saved!")

# Run the video processing
if __name__ == "__main__":
    process_video()
```
